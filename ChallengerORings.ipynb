```python
import pandas as pd

# Load the training data (sat.trn)
data = pd.read_csv("challenger.data", sep=" ", header=None)

column_names = ["N. of O-Rings","N. exp. thermal distress", "Launch temp", "Leak-check pressure", "Temporal order of flight","extra 1", "extra 2"] 

data.columns = column_names

# Display the first few rows of the DataFrame
print(data)
```

```python
print(data.describe())
```

```python
# Available Columns "N. of O-Rings","N. exp. thermal distress", "Launch temp", "Leak-check pressure", "Temporal order of flight"
temp_sorted_data = data.sort_values("Launch temp")
temp_sorted_data.plot.bar(y="N. exp. thermal distress",x="Launch temp",figsize=[14,5])
```
